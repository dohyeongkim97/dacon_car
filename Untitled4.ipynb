```python
stds = {}
for i in dflist:
    if i not in stds:
        stds[i] = int(df[i].std())

def stdpreprocessing(df):
    for i in dflist:
        mean = df[i].mean()
        for j in range(len(df)):
            if df.loc[j, i] > mean + 3*stds[i]:
                df.loc[j, i] = mean + 3*stds[i]
                
            elif df.loc[j, i] < mean - 3*stds[i]:
                df.loc[j, i] = mean - 3*stds[i]
```

```python
def scaler(df, trainset=True):
    if trainset==True:
        global manu_year
        manu_year = MinMaxScaler()
        manu_year = manu_year.fit(df[['생산년도']])
        df[['생산년도']] = manu_year.transform(df[['생산년도']])
        
        global model_year
        model_year = MinMaxScaler()
        model_year = model_year.fit(df[['모델출시년도']])
        df[['모델출시년도']] = model_year.transform(df[['모델출시년도']])
        
        global distances
        distances = StandardScaler()
        distances = distances.fit(df[['주행거리']])
        df[['주행거리']] = distances.transform(df[['주행거리']])
        
        global displ
        displ = StandardScaler()
        displ = displ.fit(df[['배기량']])
        df[['배기량']] = displ.transform(df[['배기량']])
        
    else:
        df[['생산년도']] = manu_year.transform(df[['생산년도']])
        
        df[['모델출시년도']] = model_year.transform(df[['모델출시년도']])
        
        df[['주행거리']] = distances.transform(df[['주행거리']])
        
        df[['배기량']] = displ.transform(df[['배기량']])
```

```python
class CustomDataset(Dataset):
    def __init__(self, Dataset, traindf = True):
        self.traindf = traindf
        self.Dataset = Dataset
        
        if traindf:
            self.target = pd.DataFrame(Dataset[Dataset.columns[-1]].values)
            self.data = pd.DataFrame(Dataset[Dataset.columns[1:-1]].values)
        else:
            self.data = pd.DataFrame(Dataset[Dataset.columns[1:]])
            
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        data = torch.FloatTensor(self.data.iloc[idx].values)
        
        if self.traindf:
            target = torch.FloatTensor(self.target.iloc[idx].values)
            return data, target
        else:
            return data
                
```
